```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
```

# Boxplot

```python
df_turbine = pd.read_csv(r'C:\Users\srika\Documents\Python Scripts\PythonForSASUsers\TURBINE.csv')
```

```python
#pandas
#Documentation: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.boxplot.html
df_turbine.boxplot(column = 'KWatts', by = 'Day',figsize=(10,10))
```

```python
# All days 
df_turbine.boxplot(column = 'KWatts',figsize=(10,10))
```

```python
#seaborn
# Documentation: https://seaborn.pydata.org/generated/seaborn.boxplot.html
fig, ax = plt.subplots(figsize=(10,10))
sns.boxplot(y='KWatts',x='Day',data=df_turbine, orient="v", linewidth=1,width=0.6,ax=ax).set_title('Boxplot grouped by Day')
```

# Correlation

```python
df_fitness = pd.read_csv(r'C:\Users\srika\Documents\Python Scripts\PythonForSASUsers\FITNESS.csv')
```

```python
#pandas
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.corr.html
df_fitness.corr(method='pearson')
```

```python
df_fitness.corr(method='spearman')
```

```python
df_fitness.corr(method='kendall')
```

```python
# Heatmap
pearson_corr = df_fitness.corr(method='pearson')
ax = sns.heatmap(
    pearson_corr, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
```
